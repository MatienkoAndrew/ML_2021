```python
import pandas as pd
import numpy as np
```

```python
# usecols = ['Тип объекта']
usecols=[1,2,5,6,7]
df = pd.read_csv('STIK_461531_001-01_etalon.csv', 
                 skiprows=[0], sep=';', encoding='windows-1251', usecols=usecols)
```

```python
df
```

```python
df['Номенклатура'] = df['Наименование'] + ' ' + df['Обозначение']
```

```python
df = df.drop(['Обозначение', 'Наименование'], axis=1)
```

```python
df
```

```python
df = df[['Номенклатура', 'Ед.изм.', 'Тип', 'Источник']]
```

```python
df
```

```python
df = df.rename(columns={'Ед.изм.': 'Классификатор единиц измерения'})
df
```

```python
y = {'Штука': 'шт', 
     'Литр': 'л', 
     'Миллиграмм': 'мг',
     'месяц': 'мес.',
     'грамм': 'г',
     'Единица': 'ед.',
     'килограмм': 'кг',
     'километр': 'км',
     'километр квадратный': 'км2',
     'метр': 'м',
     'метр квадратный': 'м2',
     'Метр кубический': 'м3',
     'милиметр': 'мм',
     'милиметр квадратный': 'мм2',
     'милиметр кубический': 'мм3',
     'киловатт':'кВт.ч',
     'комплект':'компол',
     'набор': 'наб.',
     'рулон': 'рул.',
     'упаковка': 'уп',
     'ящик': 'ящ'
    }
df['Классификатор единиц измерения'] = df['Классификатор единиц измерения']\
                    .apply(lambda x: x.replace(x, str(y[x])) if x in y.keys() else 'шт')
```

```python
df
```

```python
df[df['Классификатор единиц измерения'] != 'шт']
```

```python
df = df.rename(columns={'Тип': 'Виды номенклатуры'})
```

```python
df
```

```python
z = {'Деталь': 'Детали', 
     'Материал': 'Материалы', 
     'Полуфабрикат': 'Полуфабрикаты',
     'Продукция': 'Продукция',
     'Прочее изделие': 'Прочие изделия',
     'Сборочная единица': 'Сборочные единицы',
     'Стандартное изделие': 'Стандартные изделия',
     'Товар': 'Товар',
     'Услуга': 'Услуга'
    }
df['Виды номенклатуры'] = df['Виды номенклатуры']\
                    .apply(lambda x: x.replace(x, str(z[x])) if x in z.keys() else 'Прочие изделия')
```

```python
df
```

```python
df = df.rename(columns={'Источник': 'Виды воспроизводства номенклатуры'})
```

```python
df
```

```python
k = {'Изготовляемое': 'Производство', 
     'Покупное': 'Покупка', 
     'Не известно': 'Покупка'}
df['Виды воспроизводства номенклатуры'] = df['Виды воспроизводства номенклатуры']\
                    .apply(lambda x: x.replace(x, str(k[x])))
```
